```python
import pandas as pd
import seaborn as sns
sns.set(color_codes = True)
```

```python
weather = pd.read_csv('weatherHistory.csv')
```

```python
weather.head()
```

```python
weather.info()
```

```python
sns.barplot(weather['Humidity'], weather['Temperature (C)'])
```

```python
sns.distplot(weather['Humidity'])
```

```python
sns.jointplot(weather['Humidity'], weather['Temperature (C)'])
```

```python
sns.jointplot(weather['Humidity'], weather['Temperature (C)'], kind = "hex")
```

```python
sns.jointplot(weather['Humidity'], weather['Temperature (C)'], kind ="kde")
```

```python
sns.pairplot(weather[['Humidity', 'Temperature (C)', 'Visibility (km)']])
```

```python
sns.stripplot(weather['Precip Type'], weather['Temperature (C)'])
```

```python
sns.stripplot(weather['Precip Type'], weather['Temperature (C)'], jitter = True)
```

```python
sns.boxplot(weather['Humidity'], weather['Temperature (C)'], hue = weather['Precip Type'])
```

```python
sns.barplot(weather['Humidity'], weather['Temperature (C)'], hue = weather['Precip Type'])
```

```python
sns.countplot(weather['Precip Type'])
```

```python
sns.pointplot(weather['Humidity'], weather['Temperature (C)'], hue = weather['Precip Type'])
```

```python
sns.lmplot(x='Humidity', y='Temperature (C)', data = weather)
```
